## Transformando Dados

```python
df_pokemons.head()
```

```python
df_pokemons[['Attack','Sp. Atk']].head()
```

```python
# Combinando os attacks e criando uma nova coluna
df_pokemons['Total_Attack'] = df_pokemons['Attack'] + df_pokemons['Sp. Atk']
df_pokemons[['Attack','Sp. Atk','Total_Attack']].head()
```

```python
df_pokemons['Total_Defense'] = df_pokemons['Defense'] + df_pokemons['Sp. Def']
df_pokemons
```

```python
df_pokemons
```

```python
# Excluindo uma coluna
df_pokemons = df_pokemons.drop(['Total_Attack', 'Total_Defense'], axis=1) # inplace = True para substituir no data frame original
```
