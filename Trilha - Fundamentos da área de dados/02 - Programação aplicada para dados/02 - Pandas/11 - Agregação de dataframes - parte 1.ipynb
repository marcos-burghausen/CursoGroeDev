#  Agregação (Group By & Aggregate)

A agregação é simplesmente a ação de olhar as estatísticas sob a perspectiva de grupos.

<img src="https://jakevdp.github.io/PythonDataScienceHandbook/figures/03.08-split-apply-combine.png" height=600 width=600>

Basicamente, os dados vão ser separados de acordo com um ou mais grupos em comum ```(etapa de split)```. Então, a partir da separação desses grupos uma função vai ser aplicada ```(apply (sum))``` em cada subconjunto. No nosso exemplo essa função de soma. No final, ele combina o resultado agrupad ```(combine)```.

<img src="https://www.softwaretestingclass.com/wp-content/uploads/2013/06/sql_group_by_with_aggregate_function.gif" height=450 width=450>


````python
df_pokemons.groupby('Type 1').aggregate({'HP': ['min', 'max'], 'Attack': ['sum', 'mean'], 'Defense': ['sum', 'mean']})
````

````python
df_pokemons.groupby('Type 1').agg({'HP': max, 'Attack': [np.sum, np.mean], 'Defense': [np.sum, np.mean]})
````

````python
# Agrupando por Type 1 e aplicando várias funções diferentes para diferentes colunas
df_pokemons.groupby('Type 1').agg({'HP': ['max', 'min'], 'Sp. Atk': [np.std, np.median]}).reset_index()#['HP','max']
````

# Pivot Table

Usado para resumir os dados.

```python
pandas.pivot_table(data, values=None, index=None, columns=None, aggfunc='mean', fill_value=None, margins=False, dropna=True, margins_name='All', observed=False)
```

````python
df = pd.read_csv('titanic_train.csv')
df.head()
````

````python
df.drop(['PassengerId','Unnamed: 0'], inplace=True, axis=1)
````

````python
df
````

Pivot table precisa de duas entradas:
* data: o dataframe utilizado.
* index: a variável que vai ser usada para agrupar os dados.
* values: colunas a serem agregadas

````python
# pd.pivot_table(data=df, index=['Sex'], aggfunc='count')
df_sex = pd.pivot_table(data=df, index=['Sex'], aggfunc='count')
df_sex
````

````python
df_sex = pd.pivot_table(data=df, values=['Age'] , index=['Sex'], aggfunc='count')
df_sex
````

````python
# usando multiindex:
table = pd.pivot_table(df, index=['Sex','Pclass_1'], values=['Age'], aggfunc='count')*100
table
````

````python
# usando multiindex:
table = pd.pivot_table(df, index=['Sex','Pclass_1'], values=['Age'], aggfunc='mean')*100
table
````

````python
# é possível usar diferentes aggfunc além do default que é o np.mean()
table = pd.pivot_table(df, index=['Sex','Pclass_1'], aggfunc={'Age':np.mean, 'Survived':np.sum})
table
````

````python
# para selecionar algumas features use o values
table = pd.pivot_table(df, index=['Sex','Pclass_1'], values=['Survived'], aggfunc=np.sum)
table
````
