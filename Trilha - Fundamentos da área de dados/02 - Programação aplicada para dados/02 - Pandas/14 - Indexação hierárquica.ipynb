# Indexação Hierárquica

A indexação hierárquica em DataFrames do Pandas, também conhecida como MultiIndex, é uma técnica que permite criar e trabalhar com múltiplos níveis de índices em um DataFrame. Isso é útil para representar e analisar dados que têm uma estrutura de múltiplas camadas ou hierarquias.

A indexação hierárquica permite que um DataFrame tenha mais de um nível de índice para suas linhas e/ou colunas. Em vez de um único índice, você pode ter um índice de nível múltiplo que pode representar uma hierarquia de categorias.

**Para que serve?**
- Organização de Dados Complexos: Permite a organização de dados que têm múltiplas dimensões ou categorias. Por exemplo, você pode organizar dados de vendas por Country, City, e Month, o que facilita a análise em diferentes níveis de agregação.
- Facilita Análises e Resumos: Com um índice hierárquico, você pode facilmente agrupar dados em diferentes níveis e calcular estatísticas de resumo, como médias, somas, e contagens, para diferentes níveis da hierarquia.
- Melhora a Legibilidade: Dados com múltiplos níveis de agrupamento são mais legíveis e intuitivos quando organizados hierarquicamente, o que facilita a análise e interpretação dos dados.
- Acesso e Manipulação Eficiente: Permite acessar e manipular dados de maneira eficiente com base em diferentes níveis de índices. Você pode selecionar subconjuntos de dados com base em diferentes níveis da hierarquia e realizar operações específicas.
