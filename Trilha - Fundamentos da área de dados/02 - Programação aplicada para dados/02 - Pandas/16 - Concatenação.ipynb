# Concatenação

```python
s1 = pd.Series([0, 1], index=['a', 'b'])
s1
```

```python
s2 = pd.Series([2, 3, 4], index=['c', 'd', 'e'])
s2
```

```python
s3 = pd.Series([5, 6], index=['f', 'g'])
s3
```

```python
pd.concat([s1, s2, s3])
```

```python
# com dataframes:
df1 = pd.DataFrame(np.arange(6).reshape(3, 2), columns=['one', 'two'])
df2 = pd.DataFrame(5 + np.arange(4).reshape(2, 2),columns=['three', 'four'])
```

```python
df1
```

```python
df2
```

```python
pd.concat([df1, df2], axis=1)
```

```python
#se for axis = 0?
pd.concat([df1, df2], axis=0)
```

```python
#usando ignore_index
```

```python
pd.concat([df1, df2], axis=0, ignore_index=True)
```

```python
# com o mesmo nome nas colunas:
df1 = pd.DataFrame(np.arange(6).reshape(3, 2), columns=['one', 'two'])
df2 = pd.DataFrame(5 + np.arange(4).reshape(2, 2),columns=['one', 'two'])
```

```python
df1
```

```python
df2
```

```python
pd.concat([df1, df2], axis=0)
```

```python
pd.concat([df1, df2], axis=1) #join='inner' ??
```

```python
df_novo = pd.concat([df1, df2], axis=0, ignore_index=True)
df_novo
```
