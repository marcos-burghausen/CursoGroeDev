# Pandas

A biblioteca Pandas é uma das ferramentas mais poderosas e amplamente utilizadas para análise e manipulação de dados em Python. Desenvolvida por Wes McKinney, Pandas fornece estruturas de dados flexíveis e eficientes que facilitam o trabalho com dados rotulados e estruturados.

**Por que Usar Pandas?**
- **Versatilidade**: Pandas é adequado para uma ampla gama de tarefas de manipulação de dados, desde análise exploratória até limpeza e transformação de dados.
- **Integração**: Trabalha bem com outras bibliotecas Python, como NumPy, Matplotlib e Scikit-Learn, tornando-o uma escolha natural para workflows da área de dados.
- **Eficiência**: Pandas é otimizado para desempenho e pode lidar com grandes volumes de dados de maneira eficiente.

# Series

A Series é o objeto base do Pandas.

````python
pd.Series([1,2,3])
````

````python
lista = [1,2,3]
````

Series do Pandas retorna um formato vertical (Coluna).

Podemos construir `Series` partindo do zero. A forma geral de fazer isso é a seguinte:

```python
pd.Series(data, index = index)
```
`index` é um argumento opcional e `data` pode ser várias coisas

````python
pd.Series(5, index = [1, 2, 6])
````

````python
# Dicionário
dict_exemplo = {2:'a',
                1:'b',
                3:'c'}
dict_exemplo
````

````python
pd.Series(dict_exemplo).reset_index(drop=True)#.sort_values('index')#.set_index('index')
````

````python
# Criando uma série
s = pd.Series([1, 3, 5, 7, 9])
print(s)
````

````python
s[2]
````

````python
# Selecionando um elemento da série
print(s[2])
````

````python
s
````

````python
# Adicionando um novo elemento à série
s[5] = 11
s
````

````python
# Removendo um elemento da série
s = s.drop(2)
s
````

## Operações Básicas com Series

````python
series1 = pd.Series([1,2,3], index=['carne', 'carneiro', 'ovelha'])
series2 = pd.Series([2,3,4], index=['ovo', 'carneiro', 'ovelha'])
````

````python
series1
````

````python
series2
````

````python
# Soma
series1 + series2
````

````python
# Divisão
series1/series2
#(series1/series2).round(2)
````

````python
# Subtração
series1 - series2
````

````python
# Multiplicação
series1*series2
````
