# Indexação, busca e filtragem

```python
data = {'Nome': ['João', 'Maria', 'Pedro'], 'Idade': [25, 30, 35], 'Cidade': ['São Paulo', 'Rio de Janeiro', 'Curitiba']}
df = pd.DataFrame(data)
df
```

**`loc`:** O loc é utilizado para selecionar dados com base nos rótulos dos índices e colunas.

```python
# Indexação por rótulo
df.loc[0, 'Nome']
```

```python
df.loc[2, 'Cidade']
```

**`iloc`:** O iloc é utilizado para selecionar dados com base na posição das linhas e colunas (números inteiros).

```python
# Indexação por Posição
df.iloc[0, 'Nome']
```

```python
df
```

```python
df.iloc[0, 2]
```

```python
df.iloc[0, 0:2]
```

```python
df.iloc[0, [0,1]]
```

```python
df.loc[0, ['Nome', 'Cidade']]
```

```python
df.loc[0, 'Nome':'Cidade']
```

```python
df['Idade'] > 25
```

```python
df
```

```python
# Indexação Booleana
df[df['Idade'] > 25]
```

```python
df[df['Cidade']=='Curitiba']
```

Podemos filtrar as linhas do data frame baseando-se em condições.

```python
#Uma condição
df_pokemons[df_pokemons['HP']>100]
```

```python
#.loc
df_pokemons.loc[df_pokemons['HP'] > 100]
```

```python
# & -> and
# | -> or
# ~ -> not
```

```python
df_pokemons.loc[(df_pokemons['HP'] > 100) & (df_pokemons['Defense']>100)]
```

```python
df_pokemons.Name
```

```python
df_pokemons['Name']
```

```python
df_pokemons['Type 1']
```

```python
# condições múltiplas ('&' equivale a 'and',  '|' equivale a 'or')
df_pokemons.loc[(df_pokemons.HP>100) & (df_pokemons.Attack<80)]
```

```python
#.loc ('&' equivale a 'and',  '|' equivale a 'or')
df_pokemons.loc[(df_pokemons['HP']>100) | (df_pokemons['Attack']<80)]
```

## Mudança Condicional

Baseando-se nas condições que desejarmos, podemos substituir os valores de células.

Exemplo: Todo Type 1 de qualquer pokémon cujo HP seja maior que 60 será "LENDÁRIO"

```python
df_pokemons.loc[df_pokemons['HP'] > 60, 'Type 1']
```

```python
# Alterando um único valor
df_pokemons.loc[df_pokemons['HP'] > 60, 'Type 1'] = 'LENDARIO'
```

```python
df_pokemons.loc[df_pokemons['HP'] > 60, 'Type 1']
```

```python
#df_pokemons.loc[df_pokemons['Type 1']=='LENDARIO'].head()
df_pokemons[df_pokemons.HP > 60]['Type 1'].value_counts()
```

```python
df_pokemons['teste'] = 'Teste'
df_pokemons
```

```python
df_pokemons.loc[df_pokemons['HP'] > 60, 'teste'] = 'LENDARIO'
```

```python
df_pokemons.teste.value_counts()
```

```python
df_pokemons.loc[df_pokemons['HP'] > 60]
```

```python
df_pokemons['Type 1'].value_counts()
```
